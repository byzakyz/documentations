```python
import streamlit as st
import os 
import gym_super_mario_bros
from nes_py.wrappers import JoypadSpace
from gym_super_mario_bros.actions import SIMPLE_MOVEMENT
from stable_baselines3 import PPO
import IPython.display as display
import ipywidgets as widgets
import cv2
import time
import subprocess
import os

from gym.wrappers import GrayScaleObservation
from stable_baselines3.common.vec_env import VecFrameStack, DummyVecEnv
from matplotlib import pyplot as plt
```

```python
supermario_version = "SuperMarioBros-v0"

env = gym_super_mario_bros.make(supermario_version)
env = JoypadSpace(env, SIMPLE_MOVEMENT)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4
fps = 24.0  # Frames per second
frame_size = (640, 480)
out = cv2.VideoWriter(f'{supermario_version}_temp.mp4', fourcc, fps, frame_size)

done = True
frames = []

for _ in range(300):
    if done:
        state = env.reset()
    action = env.action_space.sample()
    state, reward, done, info = env.step(action)
    env.render()
    
    frame = cv2.resize(state, frame_size)
    out.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))

out.release()

current_dir = os.getcwd()

input_file = os.path.join(current_dir, f"{supermario_version}_temp.mp4")
output_file = os.path.join(current_dir, f"{supermario_version}.mp4")

try:
    os.remove(output_file)
except:
    pass

ffmpeg_command = [
    "ffmpeg",
    "-i", input_file,
    "-c:v", "libx264",
    "-crf", "23",
    "-preset", "medium",
    "-c:a", "aac",
    "-b:a", "128k",
    output_file
]

subprocess.run(ffmpeg_command)

try:
    os.remove(input_file)
except:
    pass
```
